```python
#Importar nuestra librería creada
import SebisDef as seb
import pandas as pd
```

```python
#Esta función nos permite leer el archivo que queramos, en este caso será un excel
mallorca=seb.leer_leSebis("Mallorca.xlsx")
mallorca.info()
```

```python
#Con esta función se pueden sustituir los valores nulos con el método "ffill"
mallorca_fili=seb.filiFili(mallorca)
mallorca_fili
```

```python
#Con esta función se pueden sustituir los valores nulos con el método "bfill"
mallorca_batu=seb.batuFili(mallorca)
mallorca_batu
```

```python
#Con esta función se pueden sustituir los valores nulos un string en concreto
mallorca_gong=seb.stringong(mallorca)
mallorca_gong
```

```python
mallorca_cuanti_nulos = mallorca.select_dtypes(include=["float64","int64","float","int"])
mallorca_cuali = mallorca.select_dtypes(include=["object","datetime","category"])
#Con esta función se pueden sustituir los valores nulos con el promedio de los valores
mallorca_cuanti = seb.meaniapolis(mallorca_cuanti_nulos)
mallorca_meani = pd.concat([mallorca_cuanti,mallorca_cuali],axis=1)
mallorca_meani
```

```python
#Con esta función se pueden sustituir los valores nulos con la media de los valores
mallorca_cuanti = seb.mediandina(mallorca_cuanti_nulos)
mallorca_medina = pd.concat([mallorca_cuanti,mallorca_cuali],axis=1)
mallorca_medina
```

```python
#Con esta función se pueden sustituir los valores nulos con un número constante
mallorca_constant=seb.constantinopla(mallorca)
mallorca_constant
```

```python
#Con esta función podemos contar cuántos valores nulos hay en nuestro archivo
cuenta_nulos_mallorca=seb.valoritosnulos(mallorca)
cuenta_nulos_mallorca
```

```python
cuenta_nulos_fili=seb.valoritosnulos(mallorca_fili)
cuenta_nulos_fili
```

```python
cuenta_nulos_medina=seb.valoritosnulos(mallorca_medina)
cuenta_nulos_medina
```

```python
cuenta_nulos_constant=seb.valoritosnulos(mallorca_constant)
cuenta_nulos_constant
```
